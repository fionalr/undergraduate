```python
# build PCA model and only Numpy can be used
class PCA(object):
    def __init__(self, lr, epoch):
        
        # Initializing Attributes
        self.lr = lr
        self.epoch = epoch

    def train(self, x, n_components = 1):
        
        # Initiatializing random state weights
        self.weight = np.random.rand(x.shape[1], n_components)
    
        for y in range(self.epoch):
            for x_batch in x:
                x_batch = np.array([x_batch])
                y = (x_batch.dot(self.weight))
                
                delta_weight2 = self.lr * y * np.transpose(x_batch) - y**2 * self.weight
                self.weight = self.weight + delta_weight2
                
    def predict(self,x):
        return x.dot(self.weight)
```

```python
# initialize and train the model
# Setting learning rate and # of epochs
my_pca = PCA(lr = 0.1, epoch = 50)
my_pca.train(wavData, 1)

# Predict function
result = my_pca.predict(wavData)

# Printing result
print('Reduced Result:')
print("(Num Data items, Num Dimensions) = ", result.shape)
```
